```python
import pandas as pd
import numpy as np
```

```python
beml_df=pd.read_csv("/content/BEML.csv")
beml_df[0:5]
```

```python
glaxo_df=pd.read_csv("/content/GLAXO.csv")
glaxo_df[0:5]
```

```python
beml_df=beml_df[['Date','Close']]
glaxo_df=glaxo_df[['Date','Close']]
```

```python
beml_df
```

```python
glaxo_df
```

```python
'''data frame have date column so that we can create a datetime index from this column date . it will be in ascending order'''
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
beml_df
```

```python
glaxo_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.plot(glaxo_df.Close);
plt.xlabel('Time');
plt.ylabel('Closeprice');
```

```python
plt.plot(beml_df.Close);
plt.xlabel('Time');
plt.ylabel('Closeprice');
```

```python
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods = 1)
beml_df['gain'] = beml_df.Close.pct_change(periods = 1)
```

```python
glaxo_df
```

```python
beml_df
```

```python
#drop first row since it NAN
glaxo_df=glaxo_df.dropna()
beml_df=beml_df.dropna()
```

```python
glaxo_df
```

```python
beml_df
```

```python
#Plot the gains
plt.figure(figsize=(8,6));
plt.plot(glaxo_df.index, glaxo_df.gain);
plt.xlabel('Time');
plt.ylabel('gain');
```

```python
sn.distplot(glaxo_df.gain,label='Glaxo')
plt.xlabel('Gain');
plt.ylabel('Density');
plt.legend();
```

```python
sn.distplot(beml_df.gain,label='Blaxo')
plt.xlabel('Gain');
plt.ylabel('Density');
plt.legend();
```

```python
print('Mean:',round(glaxo_df.gain.mean(),4))
print('Standard Devviation:',round(glaxo_df.gain.std(),4))
```

```python
print('Mean:',round(beml_df.gain.mean(),4))
print('Standard Devviation:',round(beml_df.gain.std(),4))
```

```python
from scipy import stats
import scipy
#Probability of making 2%loss or higher in glaxo
stats.norm.cdf(-0.02,
loc=glaxo_df.gain.mean(),
scale=glaxo_df.gain.std())
```

```python
#Probability of making 2%gain or higher in glaxo
1- stats.norm.cdf( 0.02,
loc=glaxo_df.gain.mean(),
scale=glaxo_df.gain.std())
```
